```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset that was scraped
df = pd.read_csv("books.csv")
# Display the first few rows of the dataset
df.head()
```

```python
# Get the shape of the dataset
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Clean the 'Price' column by removing unwanted characters and converting to float
df["Price"] = (
    df["Price"]
    .str.replace("Â", "", regex=False)
    .str.replace("£", "", regex=False)
    .astype(float)
)
```

```python
# Get descriptive statistics for the 'Price' column
df["Price"].describe()
```

```python
# Visualize the distribution of book prices
import matplotlib.pyplot as plt

plt.hist(df["Price"], bins=20)
plt.xlabel("Book Price (£)")
plt.ylabel("Count")
plt.title("Distribution of Book Prices")
plt.show()
```

```python
# Save the cleaned dataset to a new CSV file
df.to_csv("cleaned_books.csv", index=False)
```
